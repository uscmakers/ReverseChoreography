```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython import display

from os.path import exists
```

```python
file_name = "parser_output/extracted_features_split_3_28.csv"
df = pd.read_csv(file_name, sep = ',')
df.head()
```

```python
num_replicas = 5
```

```python
upsampleDf = df

for i in range(num_replicas):
    dfTemp = df
    randoms = pd.DataFrame(np.random.rand(*df.shape)/100)
    randoms.columns = df.columns.values
    dfTemp = dfTemp.add(randoms)
    upsampleDf = upsampleDf.append(dfTemp, ignore_index=True)
```

```python
print(upsampleDf)
```

```python
upsampleDf.to_csv("parser_output/extracted_features_split_3_28_upsample.csv")
```
