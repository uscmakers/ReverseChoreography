```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("parser_output/extracted_features.csv")
df
```

```python
df.columns
```

```python
index_no = df.columns.get_loc('RIGHT_ANKLE_DIST_z__cid_ce__normalize_True')

xAll = df.iloc[:, 1:index_no+1]
yAll = df.iloc[:, index_no+1:]
```

```python
xAll
```

### Creating Test and Training Data Sets for All Y Target Columns

```python
xTrain, xTest, yAllTrain, yAllTest = train_test_split(xAll, yAll, test_size=0.5, random_state=42)
```

```python
# defining a linear model for all y target columns
model = LinearRegression()

reg = model.fit(xTrain, yAllTrain)

# make a prediction
row = xAll.iloc[1]
yhat = model.predict([row])
# summarize prediction
print(yhat[0], '\n')

print(reg.score(xTest, yAllTest))
```

### Creating Test and Training Data Sets for a Subset of "Useful"/Key Y Target Columns

```python
print(yAll)
ySub = yAll[['danceability', 'energy', 'loudness', 'tempo']]
xTrain, xTest, yKeyComponentsTrain, yKeyComponentsTest = train_test_split(xAll, ySub, test_size=0.33, random_state=42)
```

```python
# defining a linear model for a subset of y target columns
model = LinearRegression()

reg = model.fit(xTrain, yKeyComponentsTrain)

# make a prediction
row = xAll.iloc[1]
yhat = model.predict([row])
# summarize prediction
print(yhat[0], '\n')

print(reg.score(xTest, yKeyComponentsTest))
```

## Neural Network Model

```python
# Normalizing data
# Note: Be cautious about feeding in the correct input training/testing sets 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(xTrain)
xTrain = scaler.transform(xTrain)
xTest = scaler.transform(xTest)
```

```python
print(xTrain.shape, xTest.shape)
print(yKeyComponentsTrain.shape, yKeyComponentsTest.shape)
```

```python
# Creating a NN Predictor
# https://scikit-learn.org/stable/modules/neural_networks_supervised.html
from sklearn.neural_network import MLPRegressor

reg = MLPRegressor(random_state=1, max_iter=500).fit(xTrain, yKeyComponentsTrain)
print(reg.score(xTest, yKeyComponentsTest))
```
