```python
def closest_value(input_lists, vals):
    arrs = [np.asarray(row) for row in input_lists]
    closests = [(np.abs(arrs[i] - vals[i])).argmin() for i in range(len(vals))]

    closestIndex = max(set(closests), key=closests.count)
    return closestIndex

trainLabels = []
for row in y_train_arr:
    rowVals = tuple(row)
    val = closest_value(arrs, rowVals)
    trainLabels.append(featuresList[val][1])
    
testLabels = []
for row in y_test_arr:
    rowVals = tuple(row)
    val = closest_value(arrs, rowVals)
    testLabels.append(featuresList[val][1])
    
```
